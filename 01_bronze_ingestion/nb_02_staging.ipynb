```python
from pyspark.sql.functions import col, trim, lower, upper, current_timestamp
 
spark.sql("CREATE SCHEMA IF NOT EXISTS staging")
 
customers = (
    spark.table("azure_blob_storage.src_customers")
    .withColumn("customer_name", trim(col("customer_name")))
    .withColumn("country", trim(col("country")))

    .withColumn("ingestion_ts", current_timestamp())
)
 
customers.write.mode("overwrite").saveAsTable("staging.stg_customers")
 
products = (
    spark.table("azure_blob_storage.src_products")
    .withColumn("product_name", trim(col("product_name")))
    .withColumn("category", trim(col("category")))
    .withColumn("brand", trim(col("brand")))
    .withColumn("ingestion_ts", current_timestamp())
)
 
products.write.mode("overwrite").saveAsTable("staging.stg_products")
 
invoices = (
    spark.table("azure_blob_storage.src_invoices")
    .withColumn("invoice_status", trim(lower(col("invoice_status"))))
    .withColumn("currency", trim(upper(col("currency"))))
    .withColumn("ingestion_ts", current_timestamp())
)
 
invoices.write.mode("overwrite").saveAsTable("staging.stg_invoices")

invoice_line_items = (
    spark.table("azure_blob_storage.src_invoice_line_items")
    .withColumn("ingestion_ts", current_timestamp())
)
 
invoice_line_items.write.mode("overwrite").saveAsTable("staging.stg_invoice_line_items")
 
payments = (
    spark.table("azure_blob_storage.src_payments")
    .withColumn("payment_method", trim(lower(col("payment_method"))))
    .withColumn("currency", trim(upper(col("currency"))))
    .withColumn("ingestion_ts", current_timestamp())
)
 
payments.write.mode("overwrite").saveAsTable("staging.stg_payments")
 
exchange_rates = (
    spark.table("azure_blob_storage.src_exchange_rates")
    .withColumn("from_currency", trim(upper(col("from_currency"))))
    .withColumn("to_currency", trim(upper(col("to_currency"))))
    .withColumn("ingestion_ts", current_timestamp())
)
 
exchange_rates.write.mode("overwrite").saveAsTable("staging.stg_exchange_rates")

regions = (
    spark.table("azure_blob_storage.src_regions")
    .withColumn("region_name", trim(lower(col("region_name"))))
    .withColumn("country", trim(col("country")))
    .withColumn("ingestion_ts", current_timestamp())
)
 
regions.write.mode("overwrite").saveAsTable("staging.stg_regions")
 
```
